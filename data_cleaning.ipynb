# Upload Data

```python
from google.colab import files

def upload(num_files):
    file_paths = []
    print("Uploaded:")
    for i in range(num_files):
        file_dict = files.upload()
        file_paths.append(io.BytesIO(file_dict[next(iter(file_dict))]))
        print(file_paths[i])
        
    return file_paths
```

```python
txt_files = upload(3)
```

```python
csv_files = upload(1)
```
